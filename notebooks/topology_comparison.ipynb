```python
from ete3 import PhyloTree
import pandas as pd
```

```python
path_to_nextstrain = "../data/nextstrain.uid.tree"
path_to_gisaid = "../data/gisaid.uid.tree"

tn = PhyloTree(path_to_nextstrain, format=1)
tg = PhyloTree(path_to_gisaid, format=9)

len(tn), len(tg)
```

```python
uids_from_nextstrain = set(tn.get_leaf_names())
uids_from_gisaid = set(tg.get_leaf_names())

len(uids_from_nextstrain), len(uids_from_gisaid)
```

```python
def filter_not_our(nodes):
    nodes = set(nodes)
    return nodes.intersection(uids_from_gisaid)
```

```python
len(filter_not_our(uids_from_nextstrain))
```

```python
def node_parent(node):
    return next(node.iter_ancestors())
```

```python
len((node_parent(node_parent(tn.search_nodes(name='EPI_ISL_582624')[0]))))
```

```python
left_cluster_root = tn.get_common_ancestor(
    ["EPI_ISL_775220", "EPI_ISL_768522"],
)
left_cluster_leaves = set(left_cluster_root.get_leaf_names())
left_cluster_leaves_filtered = filter_not_our(left_cluster_leaves)
print(len(left_cluster_leaves_filtered))


# Name: EPI_ISL_824286
# Name: EPI_ISL_794614
up_cluster_root = tn.get_common_ancestor(
    ["EPI_ISL_824286", "EPI_ISL_794614"],
)
up_cluster_leaves = set(up_cluster_root.get_leaf_names())
up_cluster_leaves_filtered = filter_not_our(up_cluster_leaves)
print(len(up_cluster_leaves_filtered))


# Name: EPI_ISL_527047
# Name: EPI_ISL_693483
up2_cluster_root = tn.get_common_ancestor(
    ["EPI_ISL_527047", "EPI_ISL_693483"],
)
up2_cluster_leaves = set(up2_cluster_root.get_leaf_names())
up2_cluster_leaves_filtered = filter_not_our(up2_cluster_leaves)
print(len(up2_cluster_leaves_filtered))


# Name: EPI_ISL_468062
# Name: EPI_ISL_660489
up3_cluster_root = tn.get_common_ancestor(
    ["EPI_ISL_468062", "EPI_ISL_660489"],
)
up3_cluster_leaves = set(up3_cluster_root.get_leaf_names())
up3_cluster_leaves_filtered = filter_not_our(up3_cluster_leaves)
print(len(up3_cluster_leaves_filtered))


# Name: EPI_ISL_707698
# Name: EPI_ISL_847865
right_cluster_root = tn.get_common_ancestor(
    ["EPI_ISL_707698", "EPI_ISL_847865"],
)
right_cluster_leaves = set(right_cluster_root.get_leaf_names())
right_cluster_leaves_filtered = filter_not_our(right_cluster_leaves)
print(len(right_cluster_leaves_filtered))


# Name: EPI_ISL_762992
# Name: EPI_ISL_850775
down_cluster_root = tn.get_common_ancestor(
    ["EPI_ISL_762992", "EPI_ISL_850775"],
)
down_cluster_leaves = set(down_cluster_root.get_leaf_names())
down_cluster_leaves_filtered = filter_not_our(down_cluster_leaves)
print(len(down_cluster_leaves_filtered))
```
