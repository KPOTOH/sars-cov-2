```python
with open("../data/coronavirus.json") as fin:
    tree = json.load(fin)
```

```python
def dfs(node, nwk=""):
    if not isinstance(node, dict):
        raise Exception('wow')
    name = node['name']

    if 'children' in node:
        nwk += "("
        for child in node['children']:
            nwk = dfs(child, nwk=nwk)
            nwk += ','
        nwk = nwk.rstrip(',')
        nwk += f"){name}"
    
    else:
        nwk += name
        
    return nwk


nwk = dfs(tree['tree'])
```
